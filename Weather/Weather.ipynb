```python

def wweather(date_array, key):
    timestr = time.strftime('%Y%m%d-%H%M%S')
    log = 'log_' + timestr + '.txt'
    print(f'Logging API calls in {log}')

    weather=[]
    with open(log, 'a+') as file:
        for day in date_array:
            date = day        
            try:
                url =  "https://api.worldweatheronline.com/premium/v1/past-weather.ashx?"
                key = key   # Check ConfigEnv.py
                city = "Raleigh"
                f= "json" #sets format to json
                tp = "24" #pull daily average        
                query_url = f"{url}key={key}&q={city}&format={f}&date={date}&includelocation=yes&tp={tp}"
                weather.append(requests.get(query_url).json())
                msg = f'Weather for {date} = success!\n'
                file.write(msg)
                print(msg)
            except:
                msg = f'Weather for {date} skipped!\n'
                file.write(msg)
                print(msg)
    with open('weather' + timestr + '.txt', 'a+') as file:
        print(f'Dumping json data to {file}')
        json.dump(weather, file)
```

```python
## Array containing dates for 2018-01-01 to 2018-12-31
key = Config.w_key
start3 = datetime.datetime(2018, 1, 1)
end3 = datetime.datetime(2018, 12, 31)
date_array_3 = pd.date_range(start3, end3)
wweather(date_array_3, key)
```

```python


weatherfile = 'weather20190116-095254.txt'

with open(weatherfile) as jsonfile:
    wdata = json.load(jsonfile)
    wdate = []
    maxtempF = []
    mintempF = []
    hum = []
    precip = []
    w_type = []
    wi_speed = []
    wi_gust = []
    for item in wdata:
        if len(item['data']) > 1:
            wdate.append(item['data']['weather'][0]['date'])
            maxtempF.append(item['data']['weather'][0]['maxtempF'])
            mintempF.append(item['data']['weather'][0]['mintempF'])
            hum.append(item['data']['weather'][0]['hourly'][0]['humidity'])
            precip.append(item['data']['weather'][0]['hourly'][0]['precipMM'])
            w_type.append(item['data']['weather'][0]['hourly'][0]['weatherDesc'][0]['value'])
            wi_speed.append(item['data']['weather'][0]['hourly'][0]['windspeedMiles'])
            wi_gust.append(item['data']['weather'][0]['hourly'][0]['WindGustMiles'])
#         else:
#             next
len(wdate)
```

```python
clean_weather = pd.DataFrame(
    {'Date': wdate,
     'Max Temp (f)': maxtempF,
     'Min Temp (f)': mintempF,
     'Humidity': hum,
     'Precipitation': precip,
     'Weather Conditions': w_type,
    })

clean_weather.head()

```

```python
clean_weather.to_csv("output/weather_test.csv", index=False, header=True)
```
