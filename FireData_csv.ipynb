```python
import time
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date
import seaborn as sbn
from scipy import stats
```

```python
firefile = './output/cleaned_fire_data.csv'
df = pd.read_csv(firefile, parse_dates=True, infer_datetime_format=True,skip_blank_lines=True, encoding='UTF-8')
df.dtypes

df.drop('GlobalID',axis = 1, inplace = True)
df.set_index('OBJECTID')
```

```python
df['dispatch_date_time'] = pd.to_datetime(df['dispatch_date_time'], unit = 's')
```

```python
df['arrive_date_time'] = df['arrive_date_time']/1000
df['cleared_date_time'] = df['cleared_date_time']/1000
```

```python
df.head()
```

```python
df['arrive_date_time'] = pd.to_datetime(df['arrive_date_time'], unit = 's')
df['cleared_date_time'] = pd.to_datetime(df['cleared_date_time'], unit = 's')
```

```python
df.to_csv("output/cleaned_formatted_fire_data.csv", index=False, header=True,)
```
